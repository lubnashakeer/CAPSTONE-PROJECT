# 2.1 Analysis of Literacy rate vs Total crimes.

```python
import matplotlib.pyplot as plt
feature1 = df['2001 - LIT']
feature2 = df['TOTAL IPC CRIMES']

# create a scatter plot
plt.scatter(feature1, feature2)
plt.xlabel('2001 - LIT')
plt.ylabel('CRIME')
plt.title('Relationship between Feature 1 and Feature 2')
plt.show()
```

- LOOKS LIKE THERE IS NO RELATIONSHIP BETWEEN LITERACY RATE AND CRIME - the level of literacy rate in the given population does not have a significant impact on the occurrence or frequency of criminal activities in the population , but to predict even more lets use other techniques 

- THIS IS A SIMPLE REPRESENTATION TO SEE HE RELATIONSHIP BETWEEN THE TWO FEATURES , NOW LETS USE PCA ALGORITHM FROM SCIKIT- LEARN TO PERFORM THIS ANALYSIS -(PCA is a popular unsupervised learning technique used to identify patterns and relationships in high-dimensional data. PCA transforms the data into a new set of orthogonal variables called principal components)

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

X = df[['2001 - LIT', 'TOTAL IPC CRIMES']] # select the features to analyze
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

plt.scatter(X_pca[:, 0], X_pca[:, 1])
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.show()
```

- in the low dimensional space defined by the principle components it seems like there is no relationship 

- according to the analysis the higher the literacy rate are generally associated with lower crime rate. this is because education and literacy can provide indivitual with the skills and knowledge necessary to secure stable emolpyment, make informed decisions and engage in activities. when people have access to education and can read and write, they are more likely to be productive members of society and less likely to turn to criminal activities, where in this given dataset it is sure that the litercy rate in india is not related to the crimes happening in india , but its predictable that there are many other factors that will be the reason which leads to the criminal activities
